```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Read Data

```python
df = pd.read_excel('item.csv')
df
```

```python
si = sns.heatmap(df[['Cost','Stock']].corr(),annot=True)
```

```python
# Split the data into training and test set
```

```python
x=df[['Cost']]
x
```

```python
y = df[['Stock']]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(x_test.shape)
```

```python
## Fitting the simple Linear Regression model to the training dataset


from sklearn.linear_model import LinearRegression
```

```python
## After importing  the class ,to create an object of the class named as " REG"

REG = LinearRegression()
```

```python
## Fit ()method to fit out simple linear regression object to the training set.

REG.fit(x_train,y_train)
```

```python
## Prediction of Test and Training set result

y_pred = REG.predict(x_test)
x_pred = REG.predict(x_train)
```

```python
print(y_pred)

```

```python
print(x_pred)
```

## Visualizing the Training set results

```python
## To assign the x and y values

plt.scatter(x_train,y_train,color="purple")
```

```python
plt.plot(x_train,x_pred,color="yellow")
plt.show()
```

```python
plt.scatter(x_train,y_train,color="pink")
plt.plot(x_train,x_pred,color="black")
plt.title("Stationary product")
plt.xlabel("Cost")
plt.ylabel("Stock")
plt.show()
```

## Visualization the Test Set results

```python
plt.scatter(x_test,y_test,color="purple")
plt.plot(x_train,x_pred,color="black")
plt.title("Stationary product")
plt.xlabel("Cost")
plt.ylabel("Stock")
plt.show()
```
